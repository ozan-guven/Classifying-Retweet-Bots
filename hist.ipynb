# Histogram of the number of Retweets

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#df = pd.read_csv("outputs/done3.csv", usecols=['lang', 'rt_ratio'])
df = pd.read_csv("new_outputs/user_ratio_20220601.csv")
#lang = df['lang'].to_numpy()
ratios = df[df['num_tweet'] == 200]['rt_ratio'].to_numpy()
```

```python
mean = ratios.mean() * 200
std = ratios.std() * 200
print(f"Mean: {mean} Retweets,\nStandard deviation: {std} Retweets")
print(f"Number of accounts: {ratios.shape[0]}")
print(f"Number of accounts with >= 80% RT rate: {ratios[ratios >= 0.8].shape[0]}, rest {ratios[ratios < 0.8].shape[0]}")
```

```python
plt.style.use('default')

plt.figure(figsize=(11,6))
plt.title('Ratio of Retweets')
plt.xlabel('Ratio of Retweets over the last 200 Tweets')
plt.ylabel('Number of accounts')

plt.hist(ratios, bins=10)
plt.xticks([0.1*x for x in range(11)])
#plt.savefig('hist1.png', dpi=300)

# plt.axvline(mean/200, color='r')
plt.show()
```
